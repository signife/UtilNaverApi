```python
import re

token_data = []
for temp in df['comment']:
    temp = re.sub(r"[^ã„±-ã…£ê°€-í£\s]","",temp)
    token_data.append(okt.morphs(temp))

token_data
```

```python
total_token = []

for temp in token_data:
    for temp1 in temp:
        if len(temp1) > 1:
            total_token.append(temp1)

total_token
```

```python
from collections import Counter

token_counter = Counter(total_token)

count_tokens = pd.DataFrame(dict(token_counter).items() , columns=['word','count'])
count_tokens.sort_values(by='count',ascending=False)
```

```python
senti_score = []
senti_list = []

for temp in count_tokens['word']:
    if len(senti_df[senti_df['word']==temp]) > 0:
        senti_score.append(int(senti_df[senti_df['word']==temp]['polarity'].values[0]))
        score = int(senti_df[senti_df['word']==temp]['polarity'].values[0])
        if score > 0:
            senti_list.append('ê¸ì •')
        elif score < 0:
            senti_list.append('ë¶€ì •')
        else:
            senti_list.append('ì¤‘ë¦½')
    else:
        senti_score.append(0)
        senti_list.append('ì¤‘ë¦½')

senti_score
```

```python
count_tokens['score'] = senti_score
count_tokens['senti'] = senti_list
count_tokens
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def senti_color(word,**kwargs):
    s = count_tokens[count_tokens['word']==word]['senti'].values[0]
    print(word,s)
    color = (0,0,0)
    if s=='ì¤‘ë¦½':
        color = (0,0,0)
    elif s=='ë¶€ì •':
        color = (255,0,0)
    elif s=='ê¸ì •':
        color = (0,0,255)

    return color

wc = WordCloud(font_path='NanumGothicBold.ttf',max_font_size=250,background_color='white',color_func=senti_color ).generate_from_frequencies(token_counter)
plt.figure(figsize=(8,8))
plt.imshow(wc)
```
