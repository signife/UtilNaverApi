```python
from konlpy.tag import Okt
t = Okt()
```

```python
import pandas as pd

# CSV 파일을 DataFrame으로 불러오기
df = pd.read_csv('배달의민족댓글.csv')

data = df['댓글'][4]
print(data)
```

```python
okt =Okt()
result = okt.nouns(data)
result
```

```python
from collections import Counter

print(Counter(result))
freq_data = dict(Counter(result))
freq_data
```

```python
# @title 중요. 스칼라값오류 해결 방법
sorted_freq = sorted(freq_data.items(), key = lambda item: item[1], reverse=True)

df = pd.DataFrame(sorted_freq)
df
```

```python
df = pd.read_csv('배달의민족댓글.csv',skiprows=3)
df.drop(['2'],axis=1,inplace=True)
df.rename(columns={'Unnamed: 1':'text'}, inplace=True)
df
```

```python
from tqdm import tqdm

total_list = []
okt = Okt()

# okt.nouns(df.iloc[0]['text'])
for temp in tqdm(df['text']):
    fdata = okt.nouns(temp)
    for temp1 in fdata:
        total_list.append(temp1)

total_list

from collections import Counter


freq_data = dict(Counter(total_list))
sorted_freq = sorted(freq_data.items(), key = lambda item: item[1], reverse=True)
sorted_freq
```

```python
import csv
with open("noun_list.txt",'w',newline = '',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(sorted_freq)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

sorted_freq = dict(sorted_freq)

path = 'NanumGothic.ttf'
word_cloud = WordCloud(font_path=path,
                       max_font_size=400, colormap='coolwarm').generate_from_frequencies(sorted_freq)




plt.figure(figsize = (7,6))
plt.imshow(word_cloud)
plt.axis('off')
plt.savefig('배달의민족클라우드.png')
plt.show()
# sorted_freq
```
