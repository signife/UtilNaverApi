```python
df = pd.read_csv('모니모댓글_10002.csv',skiprows=1)
df = df.rename(columns={"Unnamed: 1": "text"})
df
```

```python
text_data = ''
for temp in df['text']:
    text_data =text_data +''+ temp
text_data
```

```python
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import re
```

```python
# tdata = '안녕하세요 ~ hey~ ^^;; 내일만나요'
# kor_data = re.sub(r"[^ㄱ-ㅣ가-힣\s]","",tdata)
# kor_data

tdata = '안녕하세요 ~ hey~ ^^;; 내일만나요'
text_data = re.sub(r"[^ㄱ-ㅣ가-힣\s]","",text_data)
text_data
```

```python
okt = Okt()
okt_data =okt.pos(text_data)
okt_data
```

```python
n_adj = []
for word,tag in okt_data:
    if tag in ['Noun','Adjective']:
        if len(word) > 1:
            n_adj.append(word)
```

```python
stop_words = '어플 젤리 니모 있습니다 입니다 있어요 있는 그냥 린지'
stop_words = set(stop_words.split(' '))
stop_words
```

```python
n_adj = [word for word in n_adj if not word in stop_words]
len(n_adj)
```

```python
n_adj
```

```python
count_data = Counter(n_adj)
c_data = count_data.most_common(100)
c_data
```

```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image
import numpy as np

mask_img = np.array(Image.open('모니모마스크2.png'))
mask_img
```

```python
single_color = 'gray'

word_cloud = WordCloud(font_path = 'NanumGothic.ttf', background_color = 'white', #max font size = 젤 큰거
                      width = 2560, height = 1440, max_font_size=200, stopwords = stop_words,
                      mask=mask_img,
                      random_state=42,
                      colormap ='Paired', color_func=lambda *args, **kwargs: single_color).generate_from_frequencies(dict(c_data))



plt.figure(figsize = (16,8))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
# 이미지 파일로 저장
plt.savefig("모니모워드클라우드.png", format='png', dpi=600,transparent=True)
plt.show()
```
