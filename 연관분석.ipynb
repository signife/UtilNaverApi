```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

#support 지지도 - 전체 거래에서 물건a가 포함되는 비율
#confidence 신뢰도 - 물건 a를 구매하는 거래중에서 b를 동시에 거래되는 비율
#lift   향상도 - 두 물건 사이 구매의 우연성. a만 사는것보다 b를사고 a를 살 확률?
```

```python
dataset = [
    ['우유','콜라','빵'],
    ['바나나','우유'],
    ['콜라'],
    ['얼음','콜라'],
    ['녹차','콜라','얼음'],
    ['콜라','얼음','우유'],
    ['콜라','바나나'],
    ['딸기','바나나','녹차'],
    ['콜라','얼음']
]
```

```python
te = TransactionEncoder()
te_result = te.fit_transform(dataset)
te_result
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_result, columns=te.columns_)
df
```

```python
# @title 지지도
apriori_result = apriori(df,use_colnames=True) #지지도 1 << 모든 사람이 그 물건을 샀다. 0.5면 반만 샀다
apriori_result
```

```python
apriori_result = apriori(df,use_colnames=True, min_support = 0.1)
apriori_result
```

```python

from mlxtend.frequent_patterns import association_rules

association_rules(apriori_result, metric = 'confidence',min_threshold=0.8) #임계값
```
